# Outliers Detection

```python
import numpy as np
```

```python
arr = np.array([2,3,4,5,6,7,8,9,12,13,16,17,23,25,27,34,201])
arr
```

```python
Q1 = np.percentile(arr,25)
Q1
```

```python
Q3 = np.percentile(arr,75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
UF = Q3 + (1.5*IQR)
UF
```

```python
LF = Q1 - (1.5*IQR)
LF
```

```python
l = []
for i in arr:
    if i >= LF  and i<= UF:
        l.append(i)
arr2 = np.array(l)        
```

```python
print(arr)
print(arr2)
```

```python
import seaborn as sns 
sns.boxplot(x = arr2)
```
